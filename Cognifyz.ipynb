##Task 3
Task: Price Range vs. Online Delivery and

**1) Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.**

```python
x1=df['Has Online delivery']
y=df['Average Cost for two']
x2=df['Has Table booking']
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.boxplot(x=x1,y=y,palette="viridis")
plt.xlabel('Online Delivery')
plt.ylabel('Average Cost for Two')
plt.title('Relationship between Price Range and Online Delivery')
plt.subplot(1,2,2)
sns.boxplot(x=x2,y=y,palette="viridis")
plt.xlabel('Table Booking')
plt.ylabel('Average Cost for Two')
plt.title('Relationship between Price Range and Table Booking')
plt.tight_layout()
plt.show()
```

**Let's check for the restaurants have average coast less than 10000**

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.boxplot(x=x1,y='Average Cost for two',data=df.loc[df['Average Cost for two']<10000],palette="viridis")
plt.title('Online Delivery by cost')
plt.subplot(1,2,2)
sns.boxplot(x=x2,y='Average Cost for two',data=df.loc[df['Average Cost for two']<10000],palette="viridis")
plt.title('Table Booking by cost')
plt.tight_layout()
plt.show()
```
